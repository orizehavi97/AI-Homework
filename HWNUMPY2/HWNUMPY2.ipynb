**EX 1**
Creating and reshaping arrays

```python
randy = np.arange(1,17)
np.reshape(randy, (4,4))
```

**EX 2**
Using arrange and reshape

```python
randy1 = np.arange(10,50,5)
print(np.reshape(randy1, (2,4)))
```

**EX 3**
Randint practice

```python
randy2 = np.random.randint(1, 100, (3,3))
print(randy2)
print(randy2.max(), randy2.min())
```

**EX 4**
Slicing practice

```python
randy3 = np.random.randint(0, 21, size=10)
print(randy3)
print(randy3[:5])
print(randy3[::2])
print(randy3[::-1])
```

**EX 5**
Slicing 2D array

```python
randy4 = np.random.randint(1, 51, size=(4, 5))
print(randy4)
print(randy4[1])
print(randy4[:, 3])
```

**EX 6**
Using argmax and argmin

```python
randy5 = np.random.randint(0, 101, 15)
print(randy5)
print(np.argmax(randy5), np.argmin(randy5))
print(randy5[randy5 % 5 == 0])
print(randy5[(randy5 < 10) | (randy5 > 50)])
```
