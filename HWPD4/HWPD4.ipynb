import numpy, pandas ,display and pyplot

```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_csv('mobile.csv')
```

**EX 1**
PANDAS exercise

```python
df["first_name"] = df["Customer Name"].apply(lambda x: x.split()[0])
display(df)
```

**EX 2**
PANDAS exercise

```python
df2 = df['Phone Number'].str[2]
df['Cellular_Provider'] = df2.replace({'0':'Pelephone', '1':'Golan', '2':'Cellcom', '3':'Motorola', '4':'Orange'})
display(df)
```

**EX 3**
PANDAS exercise

```python
# A
display(df['Cellular_Provider'].value_counts())
# B
display(df['Cellular_Provider'].unique())
# C
display(df['Cellular_Provider'].nunique())
```

**EX 4**
PANDAS exercise

```python
df['User_Type'] = df.apply(lambda row: 'talker' if row['Talked'] > row['Internet'] else 'surfer',axis=1)
df['User_Type2'] = np.where(df['Talked'] > df['Internet'],'talker','surfer')
display(df)
```

**EX 5**
PANDAS exercise

```python
df3 = df.sort_values(by='Talked', ascending=True)
display(df3)
df4 = df.sort_values(by=['Talked','Internet'], ascending=[True, True])
display(df4)
```

**EX 6**
PANDAS exercise

```python
display(df.nlargest(5,'Talked'))
display(df.nsmallest(3, 'Internet'))
```
