import numpy, pandas ,display and pyplot

```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_csv("sport_athletes.csv")
display(df)
```

**EX 1**
PANDAS exercise

```python
display(df.groupby('Sport Type')[['Weight(kg)']].min().rename(columns={'Weight(kg)':'max_weight'}))
display(df.groupby('Sport Type')[['Height(cm)']].max().rename(columns={'Height(cm)':'max_height'}))
display(df.groupby('Sport Type')[['Wins']].max().rename(columns={'Wins':'max_wins'}))
```

**EX 2**
PANDAS exercise

```python
display(df.groupby('Sport Type')['Height(cm)'].nlargest(2))
```

**EX 3**
PANDAS exercise

```python
display(df.groupby(['Sport Type', 'Country']).max(numeric_only=True))
display(df.groupby(['Sport Type', 'Country']).max(numeric_only=True).loc['Basketball','USA'])
display(df.groupby(['Sport Type', 'Country']).max(numeric_only=True).xs(level='Sport Type', key='Soccer'))
display(df[df['Sport Type'].isin(['Soccer', 'Tennis'])].groupby(['Sport Type']).max())
display(df.groupby(['Sport Type', 'Country']).max().swaplevel().sort_index(level=['Sport Type', 'Country']))
display(df.agg({'Wins': ['max', 'mean', 'min', 'std'], 'Height(cm)': ['std', 'mean', 'min', 'max']}).T)
```

**EX 4**
PANDAS exercise

```python
new_rows = pd.DataFrame({
    'Country': ['Canada', 'USA'],
    'Sport Type': ['Hockey', 'Skiing'],
    'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'],
    'Height(cm)': [183, 172],
    'Weight(kg)': [84, 68],
    'Wins': [3, 4],
    'Olympic Date': ['01/02/2002', '12/02/2010']
}, index=[36, 37])
df3 = pd.concat([df, new_rows], axis=0)
display(pd.concat([df, new_rows], axis=0))
```

```python
df2 = pd.read_csv("athletes_smoking.csv")
display(df3)
display(df2)
display(pd.merge(df3, df2, on='Athlete Name', how='inner'))
display(pd.merge(df3, df2, on='Athlete Name', how='outer'))
```
