## Homework MATPLOTLIB

import numpy, pandas ,display and pyplot

```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

**EX 1**
MATPLOTLIB exercise

```python
years = [2015, 2016, 2017, 2018, 2019]
height = [110, 115, 120, 125, 132]

fig = plt.figure(figsize=(7,2))
ax1 = fig.add_axes([0, 0, 1, 1]) 
ax1.plot(years, height, label='kid height over years', color='green')
ax1.set_title('height / years')
ax1.set_xlabel('years')
ax1.set_ylabel('height')
ax1.legend()
ax1.grid(True) 

plt.show() 
```

**EX 2**
MATPLOTLIB exercise

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
temperature = [12, 14, 17, 21, 25]
rain = [80, 60, 55, 40, 20]

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

ax[0].plot(months, temperature, label='temperature over months', color='blue')
ax[0].set_title('temperature / months')
ax[0].set_xlabel('months')
ax[0].set_ylabel('temperature')
ax[0].legend()
ax[0].grid(True)

ax[1].plot(months, rain, label='rain over months', color='red')
ax[1].set_title('rain / months')
ax[1].set_xlabel('months')
ax[1].set_ylabel('rain')
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
