import numpy, pandas ,display and pyplot

```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
#Read and clean all tables
students = pd.read_csv("students.csv").dropna().drop_duplicates()
teachers = pd.read_csv("teachers.csv").dropna().drop_duplicates()
courses = pd.read_csv("courses.csv").dropna().drop_duplicates()
student_course = pd.read_csv("student_course.csv").dropna().drop_duplicates()
```

**EX 1**
test practice

```python
display(students['gender'].value_counts())

fig = plt.figure(figsize=(7,2))
ax1 = fig.add_axes([0, 0, 1, 1]) 
ax1.bar(students['gender'].unique(), students['gender'].value_counts(), color='g')
ax1.set_title('Gender count')

plt.show() 
```

**EX 2**
test practice

```python
fig = plt.figure(figsize=(7,2))
ax1 = fig.add_axes([0, 0, 1, 1]) 
ax1.hist(students['age'], bins=10, color='b', alpha=0.7)
ax1.set_title('Student Age Variation')

plt.show() 
```
