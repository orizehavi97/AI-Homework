# Question 1

##### x + y = 4
##### 2x - y = 2
#####
##### y = -x + 4
##### y = 2x - 2
#####
##### x = 2
##### y = 2
#####
##### פתרון יחיד


```python
def linear_function1(x):
   # y = -x + 4
    return -1 * x + 4

def linear_function2(x):
    # y = 2x - 2
    return 2 * x - 2


x_values = np.linspace(-15, 20, 35)
y_values1 = linear_function1(x_values)  
y_values2 = linear_function2(x_values)  
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values1, color="green", label="y = -x + 4")
plt.plot(x_values, y_values2, color="red", label="y = 2x - 2")
plt.scatter(2, 2, color='blue', zorder=5)
plt.annotate(f"(2, 2)",
             (2, 2),
             textcoords="offset points",
             xytext=(0,20),
             ha='center',
             fontsize=10,
             color='blue')
plt.legend()

plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.show()
```

# Question 2

##### 2x + y = 6
##### 4x + 2y = 10
#####
##### y = -2x + 6
##### y = -2x + 5
#####
##### אין פתרון

```python
def linear_function1(x):
   # y = -2x + 6
    return -2 * x + 6

def linear_function2(x):
    # y = -2x + 5
    return -2 * x + 5


x_values = np.linspace(-15, 20, 35)
y_values1 = linear_function1(x_values)  
y_values2 = linear_function2(x_values)  
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values1, color="green", label="y = -2x + 6")
plt.plot(x_values, y_values2, color="red", label="y = -2x + 5")
plt.legend()

plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.show()
```
